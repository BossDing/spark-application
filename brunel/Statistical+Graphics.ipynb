```python
import brunel 
```

## 夏季温度的直方图分布

```python
%brunel data('http://brunel.mybluemix.net/sample_data/US States.csv') bar x(Summer) y(#count)  bin(Summer) 
```

## 矩形区域的威士忌分级与年龄的分布

```python
%brunel data('http://brunel.mybluemix.net/sample_data/whiskey.csv') point x(Rating) y(Age) bin(Rating,Age) opacity(#count)
```

## 线性拟合夏季与美国各州冬季温度的线性拟合线

```python
%brunel data('http://brunel.mybluemix.net/sample_data/US States.csv') point x(Summer) y(Winter) + data('http://brunel.mybluemix.net/sample_data/US States.csv') line x(Summer) y(Winter) fit(Winter)
```

## 流畅的威士忌等级与价格与平滑的拟合线

```python
%brunel data('http://brunel.mybluemix.net/sample_data/whiskey.csv') line x(Rating) y(Price) smooth(Price) + data('http://brunel.mybluemix.net/sample_data/whiskey.csv') point x(Rating) y(Price)
```

## 按年龄和酒精含量过滤的条形图：平均威士忌价格

```python
%brunel data('http://brunel.mybluemix.net/sample_data/whiskey.csv') x(Category) y(Price:linear) mean(Price) bar  filter(Age, ABV)
```

## 每个地区夏季温度的中值和内四分位数范围??

```python
%brunel data('http://brunel.mybluemix.net/sample_data/US States.csv') bar x(Region) yrange(Summer) iqr(Summer) + data('sample:US States.csv') point x(Region) y(Summer) median(Summer) style('symbol:rect;height:5')
```

## 根据年龄和酒精含量(鼠标经过事件)来刷/链接威士忌价格

```python
%brunel data('http://brunel.mybluemix.net/sample_data/whiskey.csv') x(Category) y(Price:linear) mean(Price) bar interaction(select:mouseover) color(#selection) | data('sample:whiskey.csv') x(ABV) y(Age) 

```

## 国家区域的多重属性：通过阴影来显示人口，点按大小和颜色显示冬季

```python
%brunel data('http://brunel.mybluemix.net/sample_data/US States.csv') map color(population:greens) key(state) | point map color(date:[black]) size(winter:400%) key(state)
```


## 热图： 每个类别和游戏力学的游戏的数量

```python
%brunel data('http://brunel.mybluemix.net/sample_data/BGG Top 2000 Games.csv') x(Categories) y(Mechanics) color(#count:red) style('symbol:rect; size:100%; stroke:none') sort(#count)
```
